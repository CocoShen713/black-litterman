```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import math
from pypfopt.black_litterman import BlackLittermanModel, market_implied_prior_returns, market_implied_risk_aversion
from pypfopt import risk_models, EfficientFrontier, objective_functions
from pypfopt.exceptions import OptimizationError
import statsmodels.api as sm
import matplotlib.dates as mdates
from matplotlib.ticker import FuncFormatter

# ---------------------- Load data ----------------------
rates_file_path = 'T-Bill Historical Rates.xlsx'
rates_df = pd.read_excel(rates_file_path)
rates_df['Date'] = pd.to_datetime(rates_df['Date'])
rates_df.set_index('Date', inplace=True)

file_path = "BL Model Data 7-24-24.xlsx"
data = pd.read_excel(file_path, header=1)
data['date'] = pd.to_datetime(data['date'])

# Use new file
views_file_path = 'Port Models Final.xlsx'
views_df = pd.read_excel(views_file_path, sheet_name=0) # explicitly take first tab
views_df['Date'] = pd.to_datetime(views_df['Date'])

# ---------------------- Loop dates ----------------------
start_date = pd.to_datetime("2023-01-03")
end_date = pd.to_datetime("2024-06-28")
loop_dates = pd.date_range(start=start_date, end=end_date, freq='D')

# ---------------------- Posterior days list ----------------------
posterior_days_list = [90, 180, 360, 720, 900] 

percent_fmt_2dp = FuncFormatter(lambda y, _: f'{y:.2%}')

for posterior_days in posterior_days_list:
    print(f"\n================= Running backtest with posterior_days = {posterior_days} =================\n")

    # ---------------------- State holders ----------------------
    previous_weights = {ticker: 0 for ticker in data['Ticker'].unique()}

    final_asset_weights = pd.DataFrame(columns=[
        'Date', 'Ticker', 'Expected Asset Posterior Return', 'Posterior Weight',
        'Market Weight', 'Asset Closing Price', 'Expected Posterior Return View',
        'Confidence Level', 'Actual Asset Daily Return'
    ])
    final_portfolio_stats = []

    # ---------------------- Backtest loop ----------------------
    for current_date in loop_dates:
        if current_date not in data['date'].values:
            continue

        prior_start_date = current_date - pd.DateOffset(days=posterior_days)
        prior_end_date = current_date - pd.DateOffset(days=1)

        prior_data = data[(data['date'] >= prior_start_date) & (data['date'] <= prior_end_date)].copy()
        if prior_data.empty:
            continue
        prior_data.set_index('date', inplace=True)

        # covariance matrix
        adj_closing_prices = prior_data.pivot(columns='Ticker', values='Adj Closing Price')
        if adj_closing_prices.dropna(how='all').empty:
            continue
        cov_matrix = risk_models.CovarianceShrinkage(adj_closing_prices).ledoit_wolf()

        # market data
        xlb_df = prior_data[prior_data['Ticker'] == 'XLB']
        sp500_close_col = 'SP500 Closing Price'
        xlb_dates = xlb_df.index.unique()
        filtered_prices_df = prior_data.loc[prior_data.index.isin(xlb_dates)]
        if filtered_prices_df.empty or sp500_close_col not in filtered_prices_df.columns:
            continue
        market_prices = filtered_prices_df[sp500_close_col].dropna()
        if market_prices.shape[0] < 2:
            continue

        # risk-free
        if current_date in rates_df.index:
            risk_free_rate = (rates_df.loc[current_date, 'Rate']) / 100
        else:
            rf_slice = rates_df.loc[:current_date, 'Rate']
            if rf_slice.empty:
                continue
            risk_free_rate = rf_slice.iloc[-1] / 100

        delta = market_implied_risk_aversion(market_prices, frequency=252, risk_free_rate=risk_free_rate)
        
        today_slice = data.loc[data['date'] == current_date, ['Ticker', 'Mcap (Billions)']]
        if today_slice.empty:
            continue
        market_caps = today_slice.set_index('Ticker')['Mcap (Billions)'].dropna()
        common_tickers = cov_matrix.columns.intersection(market_caps.index)
        if len(common_tickers) == 0:
            continue
        market_caps = market_caps.loc[common_tickers]
        cov_matrix = cov_matrix.loc[common_tickers, common_tickers]

        prior_returns = market_implied_prior_returns(market_caps, delta, cov_matrix, risk_free_rate=risk_free_rate)

        daily_returns = data.loc[(data['date'] == current_date), ['Ticker', 'Ticker Daily Return']].set_index('Ticker')['Ticker Daily Return']
        daily_returns = daily_returns.reindex(common_tickers).fillna(0.0)
        market_weights = market_caps / market_caps.sum()

        min_weights = {}
        if current_date != loop_dates[0] and not final_asset_weights.empty:
            for ticker in market_weights.index:
                prev_w = previous_weights.get(ticker, 0.0)
                min_weights[ticker] = 0.2 * prev_w
        else:
            for ticker in market_weights.index:
                min_weights[ticker] = previous_weights.get(ticker, 0.0)
        '''
        current_views = views_df[views_df['Date'] == current_date]
        viewdict = current_views.set_index('Sector')['Expected Return'].to_dict()
        conf_series = current_views.set_index('Sector')['Confidence Level'] if not current_views.empty else pd.Series(dtype=float)

        # BL model
        bl = BlackLittermanModel(
            cov_matrix,
            pi=prior_returns,
            absolute_views=viewdict if len(viewdict) > 0 else None,
            omega='idzorek' if len(viewdict) > 0 else None,
            view_confidences=np.array([conf_series.get(k, np.nan) for k in viewdict.keys()]) if len(viewdict) > 0 else None
        )
        '''
        current_views = views_df[views_df['Date'] == current_date]
        viewdict = current_views.set_index('Sector')['Expected Return'].to_dict()
        confidences = current_views.set_index('Sector')['Confidence Level'].to_list()

        #  Filter out invalid tickers (views not in the covariance universe)
        valid_tickers = set(cov_matrix.columns)  # assets in current universe
        viewdict = {k: v for k, v in viewdict.items() if k in valid_tickers}
        conf_series = conf_series[conf_series.index.isin(valid_tickers)]

        bl = BlackLittermanModel(
            cov_matrix,
            pi=prior_returns,
            absolute_views=viewdict,
            omega='idzorek',
            view_confidences=np.array(confidences)
        )
        posterior_returns = bl.bl_returns()

        # optimization
        ef = EfficientFrontier(posterior_returns, cov_matrix)
        ef.add_objective(objective_functions.L2_reg)
        ef.add_constraint(lambda x: x >= pd.Series(min_weights))
        try:
            if float(np.max(posterior_returns)) > risk_free_rate:
                ef.max_sharpe(risk_free_rate=risk_free_rate)
            else:
                ef.min_volatility()
        except OptimizationError:
            ef = EfficientFrontier(posterior_returns, cov_matrix)
            ef.add_objective(objective_functions.L2_reg)
            ef.add_constraint(lambda x: x >= pd.Series(min_weights))
            ef.min_volatility()

        bl_weights = ef.clean_weights()
        expected_return, volatility, sharpe_ratio = ef.portfolio_performance(risk_free_rate=risk_free_rate)

        previous_weights = market_weights.to_dict()
        daily_portfolio_return = pd.Series(bl_weights).reindex(daily_returns.index).fillna(0.0).dot(daily_returns)

        sp500_day = data.loc[data['date'] == current_date, 'SP500 Daily Return']
        if sp500_day.empty:
            continue
        sp500_daily_ret = sp500_day.iloc[0]

        for ticker in bl_weights.keys():
            view_val = viewdict.get(ticker, np.nan)
            conf_val = np.nan
            if ticker in viewdict:
                conf_keys = list(viewdict.keys())
                idx = conf_keys.index(ticker)
                conf_vals = [conf_series.get(k, np.nan) for k in conf_keys]
                conf_val = conf_vals[idx]
            new_row = pd.DataFrame({
                'Date': [current_date],
                'Ticker': [ticker],
                'Expected Asset Posterior Return': [posterior_returns.get(ticker, np.nan)],
                'Posterior Weight': [bl_weights[ticker]],
                'Market Weight': [market_weights.get(ticker, np.nan)],
                'Asset Closing Price': [data.loc[(data['date'] == current_date) & (data['Ticker'] == ticker), 'Adj Closing Price'].iloc[0]
                                        if not data.loc[(data['date'] == current_date) & (data['Ticker'] == ticker), 'Adj Closing Price'].empty else np.nan],
                'Expected Posterior Return View': [view_val],
                'Confidence Level': [conf_val],
                'Actual Asset Daily Return': [daily_returns.get(ticker, np.nan)]
            })
            final_asset_weights = pd.concat([final_asset_weights, new_row], ignore_index=True)

        final_portfolio_stats.append({
            'Date': current_date,
            'Daily Risk-Free Rates': ((1 + risk_free_rate) ** (1 / 252) - 1),
            'Daily Portfolio Return': daily_portfolio_return,
            'Daily Market (S&P 500) Return': sp500_daily_ret,
            'Daily Excess Return': daily_portfolio_return - ((1 + risk_free_rate) ** (1 / 252) - 1),
            'Sharpe Ratio': sharpe_ratio,
            'Volatility': volatility
        })

    # ---------------------- Post-processing ----------------------
    if len(final_portfolio_stats) == 0 or final_asset_weights.empty:
        print(f"[posterior_days={posterior_days}] No results. Skipping outputs/plots.\n")
        continue

    final_asset_weights_df = pd.DataFrame(final_asset_weights)
    final_portfolio_stats_df = pd.DataFrame(final_portfolio_stats)

    rf_daily_series = final_portfolio_stats_df['Daily Risk-Free Rates']
    excess_ret = final_portfolio_stats_df['Daily Portfolio Return'] - rf_daily_series

    cum_portfolio_return = (1 + final_portfolio_stats_df['Daily Portfolio Return']).prod() - 1
    cum_market_return = (1 + final_portfolio_stats_df['Daily Market (S&P 500) Return']).prod() - 1

    # Sharpe using excess return and std
    mean_excess = excess_ret.mean()
    std_excess  = excess_ret.std(ddof=1)
    portfolio_sharpe_ratio = (mean_excess / std_excess) * np.sqrt(252) if std_excess not in [0, 0.0, np.nan] else np.nan
    portfolio_volatility = final_portfolio_stats_df['Daily Portfolio Return'].std(ddof=1) * np.sqrt(252)

    # ---------------------- Alpha/Beta (use excess returns) ----------------------
    df = final_portfolio_stats_df.dropna(
        subset=['Daily Portfolio Return', 'Daily Market (S&P 500) Return', 'Daily Risk-Free Rates']
    )

    portfolio_returns = df['Daily Portfolio Return']
    market_returns = df['Daily Market (S&P 500) Return']
    risk_free_rates = df['Daily Risk-Free Rates']

    # Excess returns
    excess_port = (portfolio_returns - risk_free_rates).rename('Excess Portfolio')
    excess_mkt  = (market_returns - risk_free_rates).rename('Excess Market')

    # CAPM on excess returns: (Rp - Rf) = alpha + beta * (Rm - Rf)
    X = sm.add_constant(excess_mkt)
    model = sm.OLS(excess_port, X, missing='drop').fit()
    beta = model.params.get('Excess Market', np.nan)
    raw_alpha = model.params.get('const', np.nan)          # Jensen's alpha (daily) under excess-return regression
    jensen_alpha = raw_alpha

    annualized_raw_alpha = ((1 + raw_alpha) ** 252 - 1) if pd.notna(raw_alpha) else np.nan
    annualized_jensen_alpha = ((1 + jensen_alpha) ** 252 - 1) if pd.notna(jensen_alpha) else np.nan

    print(final_asset_weights_df)
    print(final_portfolio_stats_df)

    print(f"[posterior_days={posterior_days}] Cumulative Portfolio Return over test period:", cum_portfolio_return, f"({cum_portfolio_return:.2%})")
    print(f"[posterior_days={posterior_days}] Cumulative Market Return over test period:", cum_market_return, f"({cum_market_return:.2%})")
    print(f"[posterior_days={posterior_days}] Sharpe Ratio of Portfolio over test period: {portfolio_sharpe_ratio:.2f}" if pd.notna(portfolio_sharpe_ratio) else f"[posterior_days={posterior_days}] Sharpe Ratio: NaN")
    print(f"[posterior_days={posterior_days}] Volatility of Portfolio over test period:", portfolio_volatility, f"({portfolio_volatility:.2%})")
    print(f"[posterior_days={posterior_days}] Beta: {beta:.3f}" if pd.notna(beta) else f"[posterior_days={posterior_days}] Beta: NaN")
    print(f"[posterior_days={posterior_days}] Alpha (Daily):", raw_alpha)
    print(f"[posterior_days={posterior_days}] Alpha (Annualized):", annualized_raw_alpha, f"({annualized_raw_alpha:.2%})" if pd.notna(annualized_raw_alpha) else "")
    print()

    # ---------------------- Save to Excel ----------------------
    final_asset_weights_path = f"./Final_Asset_Weights_{posterior_days}d.xlsx"
    final_portfolio_stats_path = f"./Final_Portfolio_Stats_{posterior_days}d.xlsx"

    final_asset_weights_df.to_excel(final_asset_weights_path, index=False)
    final_portfolio_stats_df.to_excel(final_portfolio_stats_path, index=False)

    summary_df = pd.DataFrame([{
        'Cumulative Portfolio Return': cum_portfolio_return,
        'Cumulative Market Return': cum_market_return,
        'Sharpe Ratio': portfolio_sharpe_ratio,
        'Volatility': portfolio_volatility,
        'Beta': beta,
        'Alpha (Daily)': raw_alpha,
        'Alpha (Annualized)': annualized_raw_alpha
    }])

    startcol = final_portfolio_stats_df.shape[1] + 2
    with pd.ExcelWriter(final_portfolio_stats_path, engine='openpyxl', mode='a', if_sheet_exists='overlay') as writer:
        summary_df.to_excel(writer, startrow=0, startcol=startcol, index=False)

    # ---------------------- Plots ----------------------
    daily_weights = final_asset_weights_df.pivot(index='Date', columns='Ticker', values='Posterior Weight')
    daily_market_weights = final_asset_weights_df.pivot(index='Date', columns='Ticker', values='Market Weight')

    etf_order = [c for c in daily_weights.columns.tolist() if c in daily_market_weights.columns]
    if len(etf_order) == 0:
        print(f"[posterior_days={posterior_days}] No overlapping tickers for plotting.\n")
        continue

    mean_market_weights = daily_market_weights[etf_order].mean()

    # Mean BL weights vs market weights
    mean_weights = daily_weights[etf_order].mean()
    print(f"[posterior_days={posterior_days}] Mean Daily Black-Litterman Allocation Weights:")
    for ticker, weight in mean_weights.items():
        print(f"{ticker}:", weight, f"({weight:.2%})")
    print()

    print(f"[posterior_days={posterior_days}] Mean Daily Market Weights:")
    for ticker, weight in mean_market_weights.items():
        print(f"{ticker}:", weight, f"({weight:.2%})")
    print()

    combined_df = mean_weights.to_frame(name='Portfolio').join(mean_market_weights.to_frame(name='Market')) * 100
    etfs = combined_df.index.tolist()
    portfolio_weights = combined_df.iloc[:, 0].values
    market_weights_vals = combined_df.iloc[:, 1].values

    x = np.arange(len(etfs)) * 1.2
    width = 0.5

    fig, ax = plt.subplots(figsize=(14, 6))
    bars1 = ax.bar(x - width / 2, portfolio_weights, width, label='Portfolio', color='royalblue')
    bars2 = ax.bar(x + width / 2, market_weights_vals, width, label='Market', color='darkorange')

    ax.set_ylabel('Allocation Mean (%)')
    ax.set_title(f'Mean Daily BL Portfolio vs Market Weights by ETF ({posterior_days}d)')
    ax.set_xticks(x)
    ax.set_xticklabels(etfs)
    ax.set_xlabel('ETF')
    ax.legend()
    ax.yaxis.grid(True, linestyle='-', alpha=0.6)

    max_val = max(np.nanmax(portfolio_weights), np.nanmax(market_weights_vals))
    y_max = np.ceil(max_val + 1)
    ax.set_ylim(0, y_max)
    ax.set_yticks(np.arange(0, y_max + 1, 2))
    ax.set_yticklabels([f"{y:.2f}%" for y in ax.get_yticks()])

    for bar in list(bars1) + list(bars2):
        height = bar.get_height()
        ax.annotate(f'{height:.2f}%',
                    xy=(bar.get_x() + bar.get_width() / 2, height),
                    xytext=(0, 2),
                    textcoords="offset points",
                    ha='center', va='bottom', fontsize=8)

    plt.tight_layout()
    plt.show()

    # Monthly average stacked bar
    monthly_avg_weights = daily_weights.resample('ME').mean()
    if not monthly_avg_weights.empty:
        fig, ax = plt.subplots(figsize=(14, 6))
        monthly_avg_weights.plot(kind='bar', stacked=True, ax=ax, colormap='tab20')
        ax.yaxis.set_major_formatter(percent_fmt_2dp)
        ax.yaxis.grid(True, linestyle='-', alpha=0.6)
        ax.set_xticklabels([d.strftime('%b %Y') for d in monthly_avg_weights.index], rotation=45)
        ax.set_title(f'Optimal BL Portfolio Allocation % - Monthly Average ({posterior_days}d)')
        ax.set_ylabel('Monthly Average Allocation (%)')
        ax.set_xlabel('Month')
        ax.legend(title='ETF', bbox_to_anchor=(1.05, 1), loc='upper left')
        plt.tight_layout()
        plt.show()

    # Std of daily weights
    weights_std = daily_weights[etf_order].std()
    weights_std = weights_std.dropna()
    if not weights_std.empty and np.isfinite(weights_std.max()):
        ceiling_std = math.ceil(float(weights_std.max()) * 100) / 100
        if ceiling_std == 0:
            ceiling_std = 0.01
        fig, ax = plt.subplots(figsize=(10, 5))
        bars = ax.bar(weights_std.index, weights_std.values, color='cornflowerblue')
        ax.set_ylim(0, ceiling_std)
        for bar in bars:
            height = bar.get_height()
            ax.annotate(f'{height:.2%}', xy=(bar.get_x() + bar.get_width() / 2, height),
                        xytext=(0, 2), textcoords='offset points', ha='center', va='bottom')
        ax.yaxis.set_major_formatter(percent_fmt_2dp)
        ax.yaxis.grid(True, linestyle='-', alpha=0.6)
        ax.set_title(f'Standard Deviation of Daily BL Allocation Weights by ETF ({posterior_days}d)')
        ax.set_ylabel('Allocation Standard Deviation (%)')
        ax.set_xlabel('ETF')
        plt.xticks(rotation=45)
        plt.tight_layout()
        plt.show()

    # Cumulative returns
    cumulative_returns = (1 + final_portfolio_stats_df[['Daily Portfolio Return', 'Daily Market (S&P 500) Return']]).cumprod()
    fig, ax = plt.subplots(figsize=(12, 6))
    ax.plot(final_portfolio_stats_df['Date'], cumulative_returns['Daily Portfolio Return'] - 1, label='Portfolio Cumulative Return')
    ax.plot(final_portfolio_stats_df['Date'], cumulative_returns['Daily Market (S&P 500) Return'] - 1, label='S&P 500 Cumulative Return', color='crimson')
    ax.yaxis.set_major_formatter(percent_fmt_2dp)
    ax.xaxis.set_major_locator(mdates.MonthLocator())
    ax.xaxis.set_major_formatter(mdates.DateFormatter('%b %d, %Y'))
    ax.set_title(f'Cumulative Returns Over Test Period: Portfolio vs. S&P 500 ({posterior_days}d)')
    ax.set_ylabel('Cumulative Growth of $1 Investment')
    ax.set_xlabel('Date')
    plt.xticks(rotation=45)
    ax.legend()
    ax.grid(True, linestyle='-', alpha=0.5)
    plt.tight_layout()
    plt.show()
```
